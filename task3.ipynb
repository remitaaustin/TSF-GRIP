#### Perform 'Exploratory Data Analysis' on dataset 'SampleSuperstore'

```python
#Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load the dataset
d=pd.read_csv("SampleSuperstore.csv")
d.head(10)
```

```python
#Information about dataset
d.info()
```

```python
#Display summary
d.describe()
```

```python
#Check for missing/null values
d.isnull().sum()
```

```python
#Check for duplicate values
d.duplicated().sum()
```

```python
#Drop duplicate values
d.drop_duplicates()
```

```python
#Drop Postal Code column which is unneccesary
d=d.drop(['Postal Code'],axis=1)
```

```python
d.head(5)
```

```python
#To find pairwise correlation of all columns
d.corr()
```

```python
fig,axes=plt.subplots(1,1,figsize=(8,8))
sns.heatmap(d.corr(),annot=True)
plt.show()
```

```python
sns.pairplot(d)
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='Region',data=d,order=d['Region'].value_counts().index)
plt.show()
```

### Inference: Most of the sales are from the Western region

```python
plt.figure(figsize=(15,10))
sns.countplot(x=d['State'],order=d['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

### Inference: California and New York shows most buys

```python
plt.figure(figsize=(10,10))
d['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

### Inference: Maximum from binders, paper, furnishings, phones etc and minimum from copiers, machines, suppliers etc

```python
plt.figure(figsize=(10,10))
plt.title('Sales in each Region wrt Segment')
sns.barplot(x='Region',y='Sales',data=d,hue='Segment',order=d['Region'].value_counts().index)
plt.xlabel('Region',fontsize=15)
plt.show()
```

### Inference: Sales in different regions wrt segment is almost the same

```python
plt.figure(figsize=(10,10))
sns.barplot(x=d['Segment'], y=d['Profit'])
```

### Inference: Profit is maximum for Home Office segment and minimum for Consumer segment

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=d)
plt.legend()
plt.show()
```

### Inference: There isn't much correlation between  discount and profit

```python
d.groupby('Category')[['Profit','Sales']].sum().plot.bar(figsize=(8,5))
plt.show()
```

### Inference: Technology have high profit and sales compared to Office Supplies and Furniture

```python
smodes=[]
salesqty=[]
for smode,data_d in d.groupby('Ship Mode'):
    smodes.append(smode)
    salesqty.append(data_d['Quantity'].sum())
smodes
```

```python
salesqty
```

### Inference: Sales quantity for the different ship modes are computed

```python
d.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
ps=d.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(figsize=(15,8))
plt.title('Profit/Loss and Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

### Inference: Phones have the highest sales wrt other sub-categories. Chairs have high sales close to phones but profit is less. Huge loss is faced by tables and bookcases sub-categories from the plot.
